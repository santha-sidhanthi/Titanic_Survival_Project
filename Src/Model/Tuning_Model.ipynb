```python
#Import Libraries


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_pickle("C:\\Users\\santh\\Titanic_Survival_Project\\Models\\EDA.pickle")

titanic_data
```

```python
titanic_data.info()
```

```python
gender=pd.get_dummies(titanic_data['Sex'],drop_first=True)

titanic_data['Gender']=gender
```

```python
titanic_data.head()
```

```python

titanic_data.columns
```

```python

titanic_data.drop(['PassengerId', 'Sex', 'Name', 'Ticket'],axis=1,inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data['Embarked'].unique()
```

```python
titanic_data['Embarked'].replace({'S':0,'C':1,'Q':2},inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data.describe()
```

```python
#Seperate Dependent and Independent variables

x=titanic_data[['Pclass','Age','SibSp','Parch','Fare','Embarked','Gender']]
y=titanic_data['Survived']
```

MODELLING


```python
#import train test split method

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
predict=lr.predict(x_test)  # Predictions
```

```python
import pickle
```

```python
with open('Train.pickle', 'wb') as f:
    pickle.dump(titanic_data, f)
```

```python
# TESTING MODEL

from sklearn.metrics import confusion_matrix
#import classification report
from sklearn.metrics import classification_report

pd.DataFrame(confusion_matrix(y_test,predict),columns=['Predicted No','Predicted Yes'],index=['Actual No','Actual Yes'])

print(classification_report(y_test,predict))
```

GRIDSEARCHCV

```python
# Grid search cross validation
from sklearn.model_selection import GridSearchCV
```

```python
#from sklearn.linear_model import LogisticRegression
grid={"C":np.logspace(2,10,100),"penalty":["l1","l2"]}# l1 lasso l2 ridge
logreg=LogisticRegression()
logreg_cv=GridSearchCV(logreg,grid,cv=10)
logreg_cv.fit(x_train,y_train)
```

```python
logreg_cv.best_params_ #this is the best parameter which our accuracy is high
```

```python
GD_predictions=logreg_cv.predict(x_test)
```

```python
print(classification_report(y_test,GD_predictions))  #MODEL TUNED WITH GRIDSEARCHCV
```

```python
#making confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, GD_predictions)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, GD_predictions)
```

```python
accuracy #model accuracy after gridsearch
```
